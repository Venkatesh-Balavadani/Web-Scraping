```python
deptime = soup.find_all('div',attrs={'class':'time'})
```

```python
deptime
```

```python
daTime = []
deptimes = []
arrtimes = []
for da in deptime:
    daTime.append(da.getText())
daTime.remove('A320 | A321')
daTime.remove('ATR')
for dept in daTime:
    deptimes.append(dept.split(" -")[0])
    arrtimes.append(dept.split("- ")[1])
```

```python
wholeDiv = soup.find_all('div',attrs={'class':'fs-wrap'})
prices
```

```python
duration = soup.find_all('span',attrs={'class':'d-md-none'})
durations = []
for dur in duration:
    durations.append(dur.getText())
price = soup.find_all('div',attrs={'class':'price'})
prices = []
for pr in price:
    prices.append(pr.getText().replace("\xa0",""))
prices.pop(0)
prices.pop(-1)
regular_price = prices[0::2]
flexi_price = prices[1::2]
flights = soup.find_all('span',attrs={'class':'flightNo'})
flight_no = []
for fn in flights:
    flight_no.append(fn.getText().replace("\xa0",""))
```

```python
totalDiv = []
for td in wholeDiv:
    totalDiv.append(td.getText())
```

```python
non_stop = []
nonStopString = []
sub = "Non-stop"
for subString in totalDiv:
    if sub in subString:
        non_stop.append(subString)
for fs in non_stop:
    start = fs.find('N')
    end = fs.find('p')
    nonStopString.append(fs[start:end+1])
```

```python
connect = []
connectString = []
sub = "Connect"
for subString1 in totalDiv:
    if sub in subString1:
        connect.append(subString1)
for fs1 in connect:
    startIndex = fs1.find('C')
    endIndex = fs1.find('t')
    connectString.append(fs1[startIndex:endIndex+1])
```

```python
flightDF = pd.DataFrame(columns=['Flight_Number','Departure','Arrival','Flight_Type','Duration','Saver','Flexi'])
```

```python
flight_type = nonStopString + connectString
```

```python
flight_no
```

```python
for index, rows in flightDF.iterrows():
    if(rows["Flight_Type"] == "Connect"):
        connectIndex = index
        connectIndex1 = connectIndex + 1
    if(index<len(flight_no)):
        index = index + 2
    flight_no[connectIndex] = flight_no[connectIndex] + " " + flight_no[connectIndex1]
    del(flight_no[connectIndex1])
```

```python
flightDF['Flight_Number'] = flight_no
flightDF['Departure'] = deptimes
flightDF['Arrival'] = arrtimes
flightDF['Flight_Type'] = flight_type
flightDF['Duration'] = durations
flightDF['Saver'] = regular_price
flightDF['Flexi'] = flexi_price
```
